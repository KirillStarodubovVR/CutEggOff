```python
import pandas as pd
import tqdm
import unicodedata

def has_accent(word):
    for char in word:
        if unicodedata.combining(char) == 230:  # Проверка символа ударения
            return True
    return False

def get_accentuated_equal(word1, word2):
    if len(word1) != len(word2):
        return False
    for char1, char2 in zip(word1, word2):
        if max(unicodedata.combining(char1), unicodedata.combining(char2)) == 0:
            if char1 != char2:
                return False
        else:
            if unicodedata.combining(char1) != unicodedata.combining(char2):
                return False
    return True
```

```python
data1 = pd.read_csv('test_1_5000.csv')
data2 = pd.read_csv('test_2_5000.csv')
```

```python
def get_accuracy(text_true, text_predicted):
    # get accentuated words
    words_true = text_true.split()
    words_predicted = text_predicted.split()
    true_cnt, all_cnt = 0, 0
    wrong_words = []
    for word_true, word_predicted in zip(words_true, words_predicted):
        if has_accent(word_true) and len(word_true) > 4:
            if get_accentuated_equal(word_true, word_predicted):
                true_cnt += 1
            else:
                wrong_words.append((word_true, word_predicted))
            all_cnt += 1
    return true_cnt, all_cnt, set(wrong_words)
```

```python
true_, all_ = 0, 0
wrong_words_set = set()
for rn, row in tqdm.tqdm(data1.iterrows()):
    pred, _ = accentuate(row['text_raw'], wordforms, lemmas)
    
    
    true_cnt, all_cnt, wrong_words = get_accuracy(row['Full context'], pred)
    wrong_words_set.update(wrong_words)
    true_ += true_cnt
    all_ += all_cnt
    
    if rn % 200 == 0:
        print(true_ / all_)
```
