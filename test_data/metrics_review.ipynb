```python
from ruaccent import RUAccent
```

```python
ru_accent = RUAccent()
ru_accent.load()
```

```python
import pandas as pd
test = pd.read_csv('test_data/test_1_5000.csv')
test
```

```python
def predict(text):
    prediction, omographs, unknown = ru_accent.process_all(text)
    return prediction


test["prediction"] = test["text_raw"].apply(lambda x: predict(x))


def omographs(text):
    prediction, omographs, unknown = ru_accent.process_all(text)
    return omographs


test["omographs"] = test["text_raw"].apply(lambda x: omographs(x))
test["prediction"] = test["prediction"].apply(lambda x: " ".join(x))
test["prediction"] = test["prediction"].apply(lambda x: x.replace("-", " "))
test["Full context"] = test["Full context"].apply(lambda x: x.replace("-", " "))
test
```

```python
accented = [
    "а́",
    "а̀",
    "я́",
    "я̀",
    "у́",
    "у̀",
    "ю́",
    "ю̀",
    "о́",
    "о̀",
    "е́",
    "ѐ",
    "э́",
    "и́",
    "ѝ",
    "ы́",
    "ы̀",
    "ё",
]
import re


def replace_accent(text):
    fin_last = []
    words = text.split(" ")

    for word in words:
        accent_index = [word.find(x) for x in accented if word.find(x) >= 0]
        if len(accent_index) == 0:
            word = word.replace("-", " ")
            word = re.sub("[^а-яёА-ЯЁ+]+", "", word)
            fin_last.append(word.lower())
        elif len(accent_index) == 1:
            word = word.replace("-", " ")
            word_new = word[: accent_index[0]] + "+" + word[accent_index[0] :]
            word_new = re.sub("[^а-яёА-ЯЁ+]+", "", word_new)
            fin_last.append(word_new.lower())
        else:
            word = word.replace("-", " ")
            word_new = word[: accent_index[0]] + "+" + word[accent_index[0] :]
            for i in range(1, len(accent_index)):
                # word = test[:accent_index[0]]+'+'+test[accent_index[0]:]
                word_new = (
                    word_new[: accent_index[i] + i]
                    + "+"
                    + word_new[accent_index[i] + i :]
                )
                word_new = re.sub("[^а-яёА-ЯЁ+]+", "", word_new)
                fin_last.append(word_new.lower())
    return fin_last
```

```python
test["true"] = test["Full context"].apply(lambda x: replace_accent(x))
test["true"] = test["true"].apply(lambda x: " ".join(x))
test["prediction"] = test["prediction"].apply(
    lambda x: re.sub("[^а-яёА-ЯЁ+ ]+", "", x)
)
test
```

```python
vowels = "аеёиоуыэюя"


def get_accuracy(text_true, text_predicted, omographs):
    # get accentuated words


    words_true = text_true.split()


    words_predicted = text_predicted.split()


    true_cnt, all_cnt = 0, 0


    wrong_words = []


    for word_true, word_predicted in zip(words_true, words_predicted):
        if (
            (sum(word_true.count(vowel) for vowel in vowels) <= 1)
            or (word_true == word_predicted)
            or (len(word_true) <= 3)
        ):
            true_cnt += 1
        elif re.sub("[^а-яёА-ЯЁ ]+", "", word_true) in [
            i.split(":")[0] for i in omographs
        ]:
            true_cnt += 1

        else:
            wrong_words.append((word_true, word_predicted))


        all_cnt += 1

    return true_cnt, all_cnt, set(wrong_words)
```

```python
print(test.iloc[427]["true"])
print(test.iloc[427]["prediction"])
print(test.iloc[427]["omographs"])
print(
    get_accuracy(
        test.iloc[427]["true"],
        test.iloc[427]["prediction"],
        test.iloc[427]["omographs"],
    )
)
```

```python
import tqdm

true_, all_ = 0, 0
wrong_words_set = set()


for rn, row in tqdm.tqdm(test.iterrows()):

    true_cnt, all_cnt, wrong_words = get_accuracy(
        row["true"], row["prediction"], row["omographs"]
    )


    wrong_words_set.update(wrong_words)


    true_ += true_cnt


    all_ += all_cnt


    if rn % 200 == 0:
        print(rn)


        print(true_ / all_)
print(f'Average accuracy {true_ / all_}')
```
